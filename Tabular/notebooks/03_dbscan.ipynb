# DBSCAN Noise Detection
Density-based clustering to label outliers as noise.

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

def detect_noise_dbscan(df, eps=0.5, min_samples=5):
    
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df)

    
    
    
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)

    
    
    labels = dbscan.fit_predict(scaled_data)

    return labels


data = {
    'age': [48, 7, 62, 48, 51, 60, 68, 11, 48, 200],
    'bp': [80, 50, 80, 70, 80, 90, 70, 80, 80, 180],
}
df = pd.DataFrame(data)

labels = detect_noise_dbscan(df, eps=1.0, min_samples=2)


noise_indices = np.where(labels == -1)[0]
print(f"Noise detected at indices: {noise_indices}")
print(f"Noisy Data:\n {df.iloc[noise_indices]}")
```
