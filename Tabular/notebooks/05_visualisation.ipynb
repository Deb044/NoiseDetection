# Visualisation for Noise Detection
This notebook combines boxplot, histogram, and scatterplot (pairplot) visual diagnostics for tabular noise analysis.

## 1. Boxplot Visualisation

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


data = np.concatenate([np.random.normal(30, 5, 200), np.random.normal(80, 10, 10), [0, 0, 0, 0]])  
df = pd.DataFrame(data, columns=['BMI'])


plt.figure(figsize=(10, 6))
sns.boxplot(x=df['BMI'], color='lightblue')
plt.title('Boxplot for Noise Detection (BMI)', fontsize=14)
plt.xlabel('BMI Value', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

## 2. Histogram Visualisation

```python
import matplotlib.pyplot as plt
import numpy as np


data = np.concatenate([np.random.normal(50, 10, 300), np.random.normal(120, 15, 10)])


mean = np.mean(data)
std = np.std(data)


plt.figure(figsize=(12, 6))
plt.hist(data, bins=40, color='gray', edgecolor='black', alpha=0.7)


plt.axvline(mean, color='blue', linestyle='--', label=f'Mean: {mean:.1f}')
for z in [-3, -2, 2, 3]:
    color = 'green' if abs(z) == 2 else 'red'
    label = f'{z} SD' if abs(z) == 3 else None  
    plt.axvline(mean + z * std, color=color, linestyle='--', label=label)

plt.title('Histogram with Z-Score Boundaries', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

## 3. Scatterplot Visualisation

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


np.random.seed(42)
n_samples = 300
age = np.random.randint(18, 90, n_samples)
glucose = np.random.normal(100, 20, n_samples)
bmi = np.random.normal(28, 5, n_samples)


glucose[0], bmi[0] = 350, 15  
glucose[1] = 0.5  
bmi[2] = 120  

df = pd.DataFrame({'Age': age, 'Glucose': glucose, 'BMI': bmi})



sns.pairplot(df, diag_kind='kde', corner=True, plot_kws={'color': 'darkblue', 'alpha': 0.6})
plt.suptitle('Multivariate Pairplot for Complex Noise Detection', fontsize=16, y=1.02)
plt.show()
```
