# 01 - Data Loading and Preview
Load grayscale images and inspect sample inputs from each folder.

```python
from pathlib import Path
import sys
import matplotlib.pyplot as plt

project_root = Path.cwd().resolve().parent
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from src.io_utils import iter_image_paths, load_image
```

```python
folders = [
    project_root / 'images' / 'random',
    project_root / 'images' / 'medical' / 'xray',
    project_root / 'images' / 'medical' / 'ultrasound',
]

for folder in folders:
    paths = iter_image_paths(folder, limit=2)
    if not paths:
        print(f'No images found in: {folder}')
        continue

    fig, axes = plt.subplots(1, len(paths), figsize=(5 * len(paths), 4))
    if len(paths) == 1:
        axes = [axes]

    for ax, p in zip(axes, paths):
        img = load_image(p)
        ax.imshow(img, cmap='gray')
        ax.set_title(p.name)
        ax.axis('off')

    plt.suptitle(f'Folder: {folder.name}', fontsize=12)
    plt.tight_layout()
    plt.show()
```
