# 02 - Noise Generation
Apply multiple synthetic noise models to one image and compare outputs.

```python
from pathlib import Path
import sys
import matplotlib.pyplot as plt

project_root = Path.cwd().resolve().parent
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from src.io_utils import iter_image_paths, load_image
from src.noise_generation import (
    add_gaussian, add_salt_pepper, add_speckle, add_poisson, add_uniform, add_mixed
)
```

```python
sample_path = iter_image_paths(project_root / 'images' / 'random', limit=2)[1]
img = load_image(sample_path)

variants = {
    'Original': img,
    'Gaussian': add_gaussian(img),
    'Salt & Pepper': add_salt_pepper(img),
    'Speckle': add_speckle(img),
    'Poisson': add_poisson(img),
    'Uniform': add_uniform(img),
    'Mixed': add_mixed(img),
}

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
for ax, (title, im) in zip(axes.flat, variants.items()):
    ax.imshow(im, cmap='gray')
    ax.set_title(title)
    ax.axis('off')

axes.flat[-1].axis('off')
plt.suptitle(f'Noise Variants for: {sample_path.name}', fontsize=12)
plt.tight_layout()
plt.show()
```
